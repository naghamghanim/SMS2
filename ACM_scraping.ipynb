```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
import time
```

```python
results = []

year=[2014,2015] #select range of years [2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2016,2016,2017,2018,2019,2020]

for y in range(len(year)):
    i=51 # search pagination start
    while i < 53: # pagination end

        try:
            # query = search query
            query = "Hate+speech+detection+OR+Sexism+detection+OR+Cyberbullying+detection+OR++Racism+detection+OR+Abusive+language+detection+OR+Offensive+language+detection%29+AND+Abstract%3A%28Hate+speech+detection+OR+Sexism+detection+OR+Cyberbullying+detection+OR++Racism+detection+OR+Abusive+language+detection+OR+Offensive+language+detection%29"
            url = 'https://dl.acm.org/action/doSearch?fillQuickSearch=false&expand=dl&AfterMonth=1&AfterYear='+str(year[y])+'&BeforeMonth=1&BeforeYear='+str(year[y]+1)+'&AllField=Title%3A%28'+query+'&pageSize=20&startPage='+str(i)
#             print(url)
            content = requests.get(url).text
            page = BeautifulSoup(content, 'lxml')

            for entry in page.find_all("div", attrs={"class": "issue-item__content"}):
                title = entry.find('h5', attrs={'class': 'issue-item__title'})
                author = entry.find('ul', attrs={'class': 'rlist--inline'})
                abst=entry.find('div', attrs={'class': 'issue-item__abstract'})
                cite=entry.find('span', attrs={'class': 'citation'})
                years=entry.find('span', attrs={'class': 'dot-separator'})
                results.append({"title": title.text.replace('[PDF]',''), "url":'https://dl.acm.org'+entry.a['href'],'authors':author.text.replace('\n',''), 'abstract':abst.text.replace('\n',''),'citation':cite.text,'years':years.text.split(',')[0].split(' ')[1]})
        except:
            print("An exception occurred")
            print(i)
            print(year[y])
        i=i+1
        
df = pd.DataFrame(results)
```
